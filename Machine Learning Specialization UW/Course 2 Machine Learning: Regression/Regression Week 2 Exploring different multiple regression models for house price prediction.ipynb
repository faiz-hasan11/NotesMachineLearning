```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dtype_dict = {'bathrooms':float, 'waterfront':int, 'sqft_above':int, 'sqft_living15':float, 'grade':int, 'yr_renovated':int, 'price':float, 'bedrooms':float, 'zipcode':str, 'long':float, 'sqft_lot15':float, 'sqft_living':float, 'floors':str, 'condition':int, 'lat':float, 'date':str, 'sqft_basement':int, 'yr_built':int, 'id':str, 'sqft_lot':int, 'view':int}

house = pd.read_csv('kc_house_data.csv', dtype = dtype_dict)
house_train = pd.read_csv('kc_house_train_data.csv', dtype = dtype_dict)
house_test = pd.read_csv('kc_house_test_data.csv', dtype = dtype_dict)

house.head()
```

# Question 1
What is the mean value (arithmetic average) of the 'bedrooms_squared' feature on TEST data? (round to 2 decimal places) 

```python
house_test['bedrooms_squared'] = house_test['bedrooms'] ** 2
house_test['bedrooms_squared'].mean()
```

# Question 2
What is the mean value (arithmetic average) of the 'bed_bath_rooms' feature on TEST data? (round to 2 decimal places) 

```python
house_test['bed_bath_rooms'] = house_test['bedrooms'] * house_test['bathrooms']
house_test['bed_bath_rooms'].mean()
```

# Question 3
What is the mean value (arithmetic average) of the 'log_sqft_living' feature on TEST data? (round to 2 decimal places) 

```python
house_test['log_sqft_living'] = house_test['sqft_living'].apply(np.log)
house_test['log_sqft_living'].mean()
```

# Question 4
What is the mean value (arithmetic average) of the 'lat_plus_long' feature on TEST data? (round to 2 decimal places) 

```python
house_test['lat_plus_long'] = house_test['lat'] + house_test['long']
house_test['lat_plus_long'].mean()
```

  * Model 1: â€˜sqft_livingâ€™, â€˜bedroomsâ€™, â€˜bathroomsâ€™, â€˜latâ€™, and â€˜longâ€™
  * Model 2: â€˜sqft_livingâ€™, â€˜bedroomsâ€™, â€˜bathroomsâ€™, â€˜latâ€™,â€˜longâ€™, and â€˜bed_bath_roomsâ€™
  * Model 3: â€˜sqft_livingâ€™, â€˜bedroomsâ€™, â€˜bathroomsâ€™, â€˜latâ€™,â€˜longâ€™, â€˜bed_bath_roomsâ€™, â€˜bedrooms_squaredâ€™, â€˜log_sqft_livingâ€™, and â€˜lat_plus_longâ€™

```python
house_train['bedrooms_squared'] = house_train['bedrooms'] ** 2
house_train['bed_bath_rooms'] = house_train['bedrooms'] * house_train['bathrooms']
house_train['log_sqft_living'] = house_train['sqft_living'].apply(np.log)
house_train['lat_plus_long'] = house_train['lat'] + house_train['long']
```

```python
model_1_input = ['sqft_living', 'bedrooms', 'bathrooms', 'lat','long']
model_2_input = ['sqft_living', 'bedrooms', 'bathrooms', 'lat','long', 'bed_bath_rooms']
model_3_input = ['sqft_living', 'bedrooms', 'bathrooms', 'lat','long', 'bed_bath_rooms', 'bedrooms_squared', 'log_sqft_living', 'lat_plus_long']
```

# Question 5
What is the sign (positive or negative) for the coefficient/weight for 'bathrooms' in model 1?

```python
from sklearn import linear_model
model_1_train = house_train[model_1_input]
model_1_target = house_train['price']
lin_reg_model_1 = linear_model.LinearRegression().fit(model_1_train, model_1_target)
print(lin_reg_model_1.coef_[model_1_input.index('bathrooms')])
```

# Question 6
What is the sign (positive or negative) for the coefficient/weight for 'bathrooms' in model 2?

```python
model_2_train = house_train[model_2_input]
model_2_target = house_train['price']
lin_reg_model_2 = linear_model.LinearRegression().fit(model_2_train, model_2_target)
print(lin_reg_model_2.coef_[model_2_input.index('bathrooms')])
```

# Question 7
Which model (1, 2 or 3) has lowest RSS on TRAINING Data?

```python
model_3_train = house_train[model_3_input]
model_3_target = house_train['price']
lin_reg_model_3 = linear_model.LinearRegression().fit(model_3_train, model_3_target)
```

```python
from sklearn.metrics import mean_squared_error
from scipy import sqrt

model_1_predict_on_train = lin_reg_model_1.predict(model_1_train)
model_2_predict_on_train = lin_reg_model_2.predict(model_2_train)
model_3_predict_on_train = lin_reg_model_3.predict(model_3_train)

print("Model 1 RSS on TRAINING Data: ", sqrt(mean_squared_error(model_1_target, model_1_predict_on_train)))
print("Model 2 RSS on TRAINING Data: ", sqrt(mean_squared_error(model_2_target, model_2_predict_on_train)))
print("Model 3 RSS on TRAINING Data: ", sqrt(mean_squared_error(model_3_target, model_3_predict_on_train)))
```

# Question 8
Which model (1, 2 or 3) has lowest RSS on TESTING Data?

```python
model_1_test = house_test[model_1_input]
model_2_test = house_test[model_2_input]
model_3_test = house_test[model_3_input]

test_target = house_test['price']

model_1_predict_on_test = lin_reg_model_1.predict(model_1_test)
model_2_predict_on_test = lin_reg_model_2.predict(model_2_test)
model_3_predict_on_test = lin_reg_model_3.predict(model_3_test)

print("Model 1 RSS on TRAINING Data: ", sqrt(mean_squared_error(test_target, model_1_predict_on_test)))
print("Model 2 RSS on TRAINING Data: ", sqrt(mean_squared_error(test_target, model_2_predict_on_test)))
print("Model 3 RSS on TRAINING Data: ", sqrt(mean_squared_error(test_target, model_3_predict_on_test)))
```
